```python
class Solution(object):
    def runningSum(self, nums):
        res = []
        total = 0
        for i in nums:
            total += i
            res.append(total)
        return res
sol = Solution()
print(sol.runningSum([1,3,5,7,9]))
```

```python
class Solution(object):
    def shuffle(self, nums, n):
        res= []
        for i in range(n):
            res.append(nums[i])
            res.append(nums[i+n])
        return res
sol = Solution()
nums = [2,5,1,3,4,7]
n = 3
print(sol.shuffle(nums, n))
```

```python
class Solution:
    def kidsWithCandies(self, candies, extraCandies):
        arr = []
        greatest = max(candies)
        for candy in candies:
            if candy + extraCandies < greatest:
                arr.append(False)
            else:
                arr.append(True)
        return arr

# Example usage:
sol = Solution()
candies = [2, 3, 5, 1, 3]
extraCandies = 3
result = sol.kidsWithCandies(candies, extraCandies)
print(result)
```
